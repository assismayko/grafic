```python
import matplotlib.pyplot as plt

# Dados
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas_2022 = [1000, 1200, 1300, 1800, 2000, 2200, 2500, 2800, 3000, 3200, 3500, 4000]
vendas_2023 = [1200, 1400, 1700, 2000, 2200, 2400, 2700, 3000, 3300, 3600, 3900, 4200]

# Criando o grfico
plt.plot(meses, vendas_2022, label='2022')
plt.plot(meses, vendas_2023, label='2023')

plt.xlabel('Ms')
plt.ylabel('Valor Total das Vendas')
plt.title('Evoluo das Vendas Mensais')
plt.legend()
plt.show()
```
